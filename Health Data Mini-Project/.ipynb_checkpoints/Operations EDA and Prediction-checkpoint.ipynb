# Korean Monthly Operations, a Predictive Analysis

In this mini-project, we will explore the number of operations done in South Korea from 2006 until 2021, and try to find relationships with the costs of said surgeries and benefits used. We hope to find strong relationships between this data, and use it to predict future operations. 

The techniques explored in this project include __basic EDA and data exploration, feature elimination, and time series prediction using auto ARIMA.__

This mini-project stems from an initial EDA performed on the gas and electricity generation, per month, in South Korea. During the feature exploration of that project, interesting trends could be found in Korean operation and cost statistics, which we will further explore here.

The original EDA can be found here: https://github.com/kevinjin21/SpringboardProjects/blob/main/Capstone/Korea%20Gas%20and%20Electricity%20Comprehensive%20EDA.ipynb

```python
operations = pd.read_excel('HealthData.xlsx', sheet_name='Operations_All_Monthly')
operations
```

We can see that our dataset ranges from the year 2006 until 2021, and has monthly data for _Number of Surgeries_, _Total Treatment Expenses_, and _Benefit Amount_. The benefits and treatment expenses are all in thousands of Won.

## Initial Visualization

For ease of use, we'll use the Month and Year columns to create a DATE column of datetime type, and use this as our index. After all, we're dealing with time series data!

```python
operations.index = pd.to_datetime(operations[['Year', 'Month']].assign(DAY=1))
operations.head()
```

Let's plot to see what kind of trends our data follow!

Just from looking at the data, there is a clear scale difference between surgery count and expenses, so we'll plot them separately for now.

```python
op_count = operations.iloc[:,2]
op_count.plot()
plt.xlabel('Time')
plt.ylabel('Number of Surgeries')
plt.title('Trends of Surgery Cases');
```

```python
costs = operations.iloc[:,3:5]
costs.plot()
plt.xlabel('Time')
plt.ylabel('Cost, Thousand Won')
plt.title('Trends of Surgery Costs');
```

Just visually, we can already see that the data is non-stationary time series data, and that they are periodic in nature. Also, the relationship between total treatment expenses and benefits amounts is very clear: the total expense is basically just benefits amount with a constant amount added on top.

## Normalize and visualize again
Since the data is on a very different scale, let's see how the periods interact by normalizing the data and plotting all together. We can see how similar the periodicity is if the data is all using the same scale. We will use MinMaxScaler for this.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler()
```

```python
scaled_data = operations.copy()
for i in range (2,5):
    scaled_data.iloc[:,i] = sc.fit_transform(scaled_data.iloc[:,i].array.reshape(-1,1))
scaled_data
```

```python
scaled_data.drop(columns=['Year','Month']).plot()
plt.xlabel('Time')
plt.ylabel('Count')
plt.title('Trends of Surgery Cases and Expenses');
```

There does seem to be some relationship between the periodicity of each feature! Especially with the expenses and benefit amounts, we can clearly see that it's the same data, just with a constant offset. After normalization, they are the same plot!

## Visualize Correlation (heatmap)
We can already tell that the features should be at least somewhat related. Let's see this in a more concrete way using a correlation heatmap.

```python
sns.heatmap(scaled_data.drop(columns=['Year','Month']).corr(), annot=True);
```

It's clear that there is strong correlation between the data. Also, we can clearly see that the total expenses and benefit amount are not independent features, so moving forward we will just use the __total expenses__ for prediction. Now that we've established this, let's try to use the expense data and past surgery data to create a model to predict future surgery cases!

# Predicting future surgery case count

It is important to make a quick note here that it is possible to do this manually, exploring the ARIMA model and finding the p, q, and d values manually. However, this is already covered in the Springboard mec-18.5.1 mini-project, so instead we will focus on using __Auto ARIMA__.

The previous mini-project can be accessed here: https://github.com/kevinjin21/mec-mini-projects/blob/master/mec-18.5.1-time-series-analysis-mini-project/Mini_Project_Time_Series_Forecasting.ipynb

```python
# install auto ARIMA if not already installed 
# !pip install pmdarima
```

```python
# import
import pmdarima as pm
```

### Initialize the auto ARIMA function
The parameters we use for auto ARIMA include:

m=12 (12 months)
<br>test='adf' (we're using Augmented Dickey-Fuller to test for stationarity)
<br>start_p,q (minimum values for p and q)
<br>max_order (maximum values for p, q, d. In this case, we'll set it to 5)

```python
# smaller df for just relevant feature and target
df = operations.iloc[:, 2:4]
df.head()
```

```python
model = pm.auto_arima(df.iloc[:,0], X=df.iloc[:,1].array.reshape(-1,1), 
                        m=12, seasonal=True,
                        start_p=0, start_q=0, max_order=5, test='adf',error_action='ignore',  
                        suppress_warnings=True, stationary=False,
                        stepwise=True)
```

```python
print(model.summary())
print(model.params)
```

Above, we can see the results of the automatically chosen p, q, d values (1, 0, 1). Let's split training and test datasets to try out our model!

### Split train and test data

```python
train_data, test_data = df.iloc[0:int(len(df)*0.8), :], df.iloc[int(len(df)*0.8):, :]
```

```python
plt.figure(figsize=(12,7))
plt.title('Total Operations')
plt.xlabel('Dates')
plt.ylabel('Cases')
plt.plot(train_data.iloc[:,0], 'blue', label='Training Data')
plt.plot(test_data.iloc[:,0], 'green', label='Testing Data')
plt.legend();
```

```python
print(test_data.shape)
```

### Fit the model and predict!

```python
model.fit(train_data.iloc[:,0], X=train_data.iloc[:,1].array.reshape(-1,1))
```

```python
forecast = model.predict(n_periods=39, X=test_data.iloc[:,1].array.reshape(-1,1))
forecast
```

```python
forecast = pd.DataFrame(forecast,index = test_data.index,columns=['Prediction'])
forecast.head()
```

### Plot Results

```python
plt.figure(figsize=(12,7))
plt.title('Predictions of Operation Count Based on Expenses')
plt.xlabel('Dates')
plt.ylabel('Operation Count')
plt.plot(train_data.iloc[:,0], color='blue', label='Training Operation Count')
plt.plot(test_data.iloc[:,0], color='green', marker='.', label='Testing Operation Count')
plt.plot(forecast, color='red', linestyle='--', label='Forecasted Operation Count')
plt.legend();
```
